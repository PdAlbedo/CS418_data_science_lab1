**Exercise 1.1** Write a python function that returns the sum of squares in range  [1, 2, ..., n ]

```python
# n = int(input('n = '))
def sum_(n):
    result = 0
    for c in range(1, n + 1):
        result += (c ** 2)

    return result
# print(sum_(n))
```

**Exercise 1.2** Write a lambda function that returns the sum of squares in range  [1, 2, ..., n ]

```python
from functools import reduce
n = int(input('n = '))
result = reduce(lambda x, y: (x) + (y ** 2), range(1, n + 1))
print(result)
```

**Exercise 2.1** For a given string s, write a function that returns a new string that doubles
each character of the original string. For instance, if the user enters _Python_, the output should be
_PPyytthhoonn_.

```python
def double_letters (s):
    # complete this function
    new_string = ""
    for ch in s:
        new_string += ch * 2
    s = new_string
    return s
    
    
s = 'Hello'
print(double_letters(s))    
```

_Example:_

```python
reverse_dict({"Alice": "Chicago", "Bob": "Detroit", "Jenna": "Chicago", "Adam":"Chicago"})
# => {"Chicago": ["Alice", "Jenna", "Adam"], "Detroit": ["Bob"]}
```

```python
from collections import defaultdict
def reverse_dict(dict_):
    reversed_dict = dict()
    for key, value in dict_.items():
        reversed_dict.setdefault(value, list()).append(key)
    print(reversed_dict)
```

**Exercise 4.4** In this exercise we will find the frequency of words in _The Yellow Wallpaper_ by _Charlotte Perkins Gilman_. The following line of code will read the entire contents of a file containing _The Yellow Wallpaper_ text and stores it in a string called _data_:


```python
dat = open('Yellow_Wallpaper_PROJECT_GUTENBERG.txt').read()
```

In order to analyze this text do the following:<br>

1. Turn the whole string to lowercase
2. Remove all the punctuations from the text (you can use "from string import punctuation")
3. Split the string into a list of individual words
4. Create a dictionary and count the frequency of words 
5. Print your dictionary in alphabetical order

```python
# 1. Turn the whole string to lowercase
dat = open('Yellow_Wallpaper_PROJECT_GUTENBERG.txt', encoding = "utf8").read()
dat = dat.lower()
print(dat)
```

```python
# 2. Remove all the punctuations from the text (you can use "from string import punctuation")
from string import punctuation
removed_str = ''.join(c for c in dat if c not in punctuation and c != '\n' and c != '—' and c != '“' and c != '”')
print(removed_str)
```

```python
import scipy.stats as st
st.norm.cdf(-1.415)
```

```python
# 3. Split the string into a list of individual words
splitted_str = removed_str.split()
print(splitted_str)
```

```python
# 4. Create a dictionary and count the frequency of words
word_fre = {}
for word in splitted_str:
    if word not in word_fre:
        word_fre[word] = 1
    else:
        word_fre[word] += 1
print(word_fre)
```

```python
# 5. Print your dictionary in alphabetical order
for key, value in sorted(word_fre.items()):
        print(key, value)
```
